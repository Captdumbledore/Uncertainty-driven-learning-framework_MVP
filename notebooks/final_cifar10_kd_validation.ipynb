## Experiment Configuration

- **Framework:** Adaptive Knowledge Reasoning Module (AKRM)
- **Dataset:** CIFAR-10

- **Learning Protocol:** Knowledge Distillation
- **MC Dropout Passes:** 30
- **Top Uncertain Samples:** 10%

## Purpose

The notebook executes the complete experimental pipeline from baseline training through uncertainty estimation, AKRM reasoning, retraining, and final evaluation.
